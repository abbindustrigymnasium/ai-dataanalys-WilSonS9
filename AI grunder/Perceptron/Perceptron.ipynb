```python
import math
import random
import tqdm
```

```python
def activationFunction(x):
    if x < 0:
        return 0
    else:
        return 1
```

```python
def f(x):
    return -5*x
f(10)
```

```python
inputs = []

for n in range(100):
    x, y = random.uniform(0, 10), random.uniform(f(0), f(10))
    # Alla på eller under linjen är 0, alla över är 1
    label = int(y > f(x))
    inputs.append((x, y, 1, label))
inputs
```

```python
class Perceptron:
    def __init__(self):
        self.lr = 0.1
        self.ratio = 1
        
        weights = []

        for n in range(len(inputs[0])-1):
            weights.append(random.random())
            
        self.weights = weights

    def activationFunction(self, x):
        if x < 0:
            return 0
        else:
            return 1
    
    def predict(self, inputs):
        sigma = 0
        i = 0
        while i < len(inputs) - 1:
            sigma += self.weights[i] * inputs[i]
            i += 1
        return self.activationFunction(sigma)
    
    def tune(self, error, inps):
        i = 0
        while i < len(inps) - 1:
            self.weights[i] += self.lr * error * inps[i]
            i += 1
            
    def train(self):
        wrongs = 0
        for inp in inputs:
            pred = self.predict(inp)
            error = inp[-1] - pred
            self.tune(error, inp)
            if inp == pred:
                wrongs += 1
        self.ratio = wrongs / len(inputs)

p = Perceptron()
```

```python
answers = []

for inp in inputs:
    answers.append((inp[-1], p.predict(inp)))

answers
```

```python
for x in tqdm.trange(10000):
    p.train()
    print(ratio)
    if p.ratio == 1.0:
        print(x)
        break
```
